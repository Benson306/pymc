<h1>Part 1: Varying-Intercept Model (Simulated Response Variable)</h2>

<h2>Q1.1 - Simulate Response</h2>

```python
# set seed for consistency across runs
np.random.seed(109)

# Set the number of districts and number of women in each district
J = 60
N = len(df_train)

# Set the fixed effects
beta_0 = np.random.normal(2, np.sqrt(1), J)
beta_1 = 4
beta_2 = -3
beta_3 = -2

# Simulate the response variable for each woman in each district
simulated_Y = np.zeros((N, J))
for j in range(J):
    p_ij = np.exp(beta_0[j] + beta_1 * df_train['urban'] + beta_2 * df_train['living.children'] + beta_3 * df_train['age_mean']) / (1 + np.exp(beta_0[j] + beta_1 * df_train['urban'] + beta_2 * df_train['living.children'] + beta_3 * df_train['age_mean']))
    simulated_Y[:, j] = np.random.binomial(1, p_ij)
    
# Compute the proportion of women who use contraceptives in each district
prop_contraceptives = np.mean(simulated_Y, axis=0)

# Print the average proportion of women who use contraceptives across all districts
print('Average proportion of women using contraceptives:', np.mean(prop_contraceptives))

# Compute the average proportion of women using contraceptives
avg_prop_contraceptives = np.mean(prop_contraceptives)

# Check if the average proportion falls within the range of 46% to 49%
if 0.46 <= avg_prop_contraceptives <= 0.49:
    print("The simulated data has an average proportion of women using contraceptives between 46% and 49%.")
else:
    print("The simulated data does not have an average proportion of women using contraceptives between 46% and 49%.")
```

<h2>Q1.2 Fit Varying-Intercept Model (Simulated Data)</h2>

```python
import pymc3 as pm

# Define the PyMC model
with pm.Model() as model:
    # Priors for fixed effects
    beta_0 = pm.Normal("beta_0", mu=0, sd=10000)
    beta_1 = pm.Normal("beta_1", mu=0, sd=10000)
    beta_2 = pm.Normal("beta_2", mu=0, sd=10000)
    beta_3 = pm.Normal("beta_3", mu=0, sd=10000)

    # Priors for varying intercepts
    mu_0 = pm.Normal("mu_0", mu=0, sd=10000, shape=60)
    sigma_0 = pm.Gamma("sigma_0", alpha=0.1, beta=0.1)

    # Expected value of outcome
    pi = pm.math.invlogit(mu_0[df_train["district"].values] + 
                          beta_1 * df_train["urban"].values + 
                          beta_2 * df_train["living.children"].values + 
                          beta_3 * df_train["age_mean"].values)

    # Likelihood of observations
    y_obs = pm.Bernoulli("y_obs", p=pi, observed=simulated_Y)

    # MCMC sampling
    trace = pm.sample(10000, tune=2000, chains=4, cores=4, target_accept=0.95)

```
